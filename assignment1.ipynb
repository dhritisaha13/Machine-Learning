```python
import pandas as pd
data=pd.read_csv('/home/dhriti/Documents/ML/breast-cancer-wisconsin-data/data.csv')
print(data)
target=data.iloc[:,1]
print(target)
df=data.iloc[:,2:32]
print(df)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
gnb = GaussianNB()
scores = cross_val_score(gnb, df, target, cv=5)
print(scores)
print("Accuracy: %.3f%% " % (scores.mean()*100.0))
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0, max_iter = 150).fit(df,target)
scores1 = cross_val_score(clf, df, target, cv=5)
print(scores1)
print("Accuracy: %.3f%%" % (scores1.mean()*100.0))
```
